```python
import os
import zipfile

local_zip = '/content/FER Dataset.zip'

zip_ref = zipfile.ZipFile(local_zip, 'r')

zip_ref.extractall('/tmp')
zip_ref.close()
```

```python
base_dir = '/tmp'

train_dir = os.path.join(base_dir, 'train')
test_dir = os.path.join(base_dir, 'test')

# Directory with our training emotion pictures
train_angry_dir = os.path.join(train_dir, 'angry')
train_disgusted_dir = os.path.join(train_dir, 'disgusted')
train_fearful_dir = os.path.join(train_dir, 'fearful')
train_happy_dir = os.path.join(train_dir, 'happy')
train_neutral_dir = os.path.join(train_dir, 'neutral')
train_sad_dir = os.path.join(train_dir, 'sad')
train_surprised_dir = os.path.join(train_dir, 'surprised')

# Directory with our testing emotion pictures
test_angry_dir = os.path.join(test_dir, 'angry')
test_disgusted_dir = os.path.join(test_dir, 'disgusted')
test_fearful_dir = os.path.join(test_dir, 'fearful')
test_happy_dir = os.path.join(test_dir, 'happy')
test_neutral_dir = os.path.join(test_dir, 'neutral')
test_sad_dir = os.path.join(test_dir, 'sad')
test_surprised_dir = os.path.join(test_dir, 'surprised')
```

```python
%matplotlib inline

import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Parameters for our graph; we'll output images in a 7x2 configuration
nrows = 7
ncols = 2

pic_index = 0 # Index for iterating over 
```

```python
# Set up matplotlib fig, and size it to fit 7x2 pics
fig = plt.gcf()
fig.set_size_inches(ncols * 7, nrows * 2)

pic_index += 2

next_angry_pix = [os.path.join(train_angry_dir, fname) for fname in train_angry_fnames[ pic_index - 2:pic_index]]
next_disgusted_pix = [os.path.join(train_disgusted_dir, fname) for fname in train_disgusted_fnames[ pic_index - 2:pic_index]]
next_fearful_pix = [os.path.join(train_fearful_dir, fname) for fname in train_fearful_fnames[ pic_index - 2:pic_index]]
next_happy_pix = [os.path.join(train_happy_dir, fname) for fname in train_happy_fnames[ pic_index - 2:pic_index]]      
next_neutral_pix = [os.path.join(train_neutral_dir, fname) for fname in train_neutral_fnames[ pic_index - 2:pic_index]]
next_sad_pix = [os.path.join(train_sad_dir, fname) for fname in train_sad_fnames[ pic_index - 2:pic_index]]
next_surprised_pix = [os.path.join(train_surprised_dir, fname) for fname in train_surprised_fnames[ pic_index - 2:pic_index]]

for i, img_path in enumerate(next_angry_pix + next_disgusted_pix + next_fearful_pix + next_happy_pix 
                             + next_neutral_pix + next_sad_pix + next_surprised_pix):
  # Set up subplot; subplot indices start at 1
  sp = plt.subplot(nrows, ncols, i + 1)
  sp.axis('Off') # Don't show axes (or gridlines)

  img = mpimg.imread(img_path)
  plt.imshow(img)

plt.show()
```

### Data Preprocessing

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# All images will be rescaled by 1./255.
train_datagen = ImageDataGenerator( rescale = 1.0/255. )
test_datagen  = ImageDataGenerator( rescale = 1.0/255. )

# --------------------
# Flow training images in batches of 20 using train_datagen generator
# --------------------
train_generator = train_datagen.flow_from_directory(train_dir,
                                                    batch_size=20,
                                                    class_mode='categorical',
                                                    color_mode = "grayscale",
                                                    target_size=(48, 48))     
# --------------------
# Flow validation images in batches of 20 using test_datagen generator
# --------------------
validation_generator =  test_datagen.flow_from_directory(test_dir,
                                                         batch_size=20,
                                                         class_mode  = 'categorical',
                                                         color_mode = "grayscale",
                                                         target_size = (48, 48))
```
